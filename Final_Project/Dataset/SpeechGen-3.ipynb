```python
import os
import random
import librosa
import numpy as np
import soundfile as sf
from TTS.api import TTS



# Initialize directories and settings
OUTPUT_DIR = "audio_output"
AUGMENTED_DIR = os.path.join(OUTPUT_DIR, "augmented")
os.makedirs(AUGMENTED_DIR, exist_ok=True)
SAMPLING_RATE = 8000  # Higher sampling rate for better quality
DURATION = 1  # seconds

# Define keywords
KEYWORDS = ["Start.", "Classic.", "Jazz.", "Rock.", "Hits."]

# Initialize TTS with a multi-speaker model
#tts = TTS(model_name="tts_models/en/multi-speaker/vctk/vits", progress_bar=False, gpu=False)
tts = TTS(model_name="tts_models/en/vctk/vits", progress_bar=False, gpu=False)

# Get all available speakers
speakers = tts.speakers  # Retrieve list of speakers from the model

print(f"Available speakers: {speakers}")

# Function to generate audio for a specific speaker
def generate_audio(keyword, speaker, output_dir):
    sanitized_keyword = keyword.replace(" ", "_").replace(".", "")
    output_path = os.path.join(output_dir, f"{sanitized_keyword}_{speaker}.wav")
    tts.tts_to_file(
        text=keyword,
        speaker=speaker,  # Use the current speaker
        file_path=output_path
    )
    print(f"Generated: {output_path}")

# Generate audio for all keywords using all speakers
def generate_all_audio():
    for keyword in KEYWORDS:
        for speaker in speakers:
            generate_audio(keyword, speaker, OUTPUT_DIR)


# Function to standardize audio length and sampling rate
def standardize_audio(audio):
    target_length = SAMPLING_RATE * DURATION
    if len(audio) < target_length:
        padding = target_length - len(audio)
        audio = np.pad(audio, (0, padding), "constant")
    else:
        audio = audio[:target_length]
    return audio.astype(np.float32)  # Ensure float32


# Function for pitch shifting
def pitch_shift(audio, sr, steps):
    return librosa.effects.pitch_shift(audio.astype(np.float32), sr=sr, n_steps=steps)


# Function for time stretching
def time_stretch(audio, rate):
    return librosa.effects.time_stretch(audio.astype(np.float32), rate=rate)


# Function for adding noise
def add_noise(audio):
    noise = np.random.normal(0, 0.005, audio.shape).astype(np.float32)
    return audio + noise


# Function for volume adjustment
def adjust_volume(audio, gain):
    return (audio * gain).astype(np.float32)


# Function for reverberation simulation
def add_reverb(audio, sr):
    delay = int(0.1 * sr)  # 100ms delay
    reverb = np.zeros(len(audio) + delay, dtype=np.float32)
    reverb[:len(audio)] += audio
    reverb[delay:] += 0.5 * audio
    return reverb[:len(audio)]


# Function to apply random augmentations
def apply_random_augmentations(audio, sr):
    augmentations = [
        lambda x: pitch_shift(x, sr, steps=random.choice([-4, 4])),
        lambda x: time_stretch(x, rate=random.choice([0.8, 1.2])),
        lambda x: add_noise(x),
        lambda x: adjust_volume(x, gain=random.choice([0.5, 1.5])),
        lambda x: add_reverb(x, sr),
    ]
    num_augmentations = random.randint(1, 3)  # Apply 1 to 3 random augmentations
    for _ in range(num_augmentations):
        augmentation = random.choice(augmentations)
        audio = augmentation(audio)
    return audio.astype(np.float32)  # Ensure float32






```
